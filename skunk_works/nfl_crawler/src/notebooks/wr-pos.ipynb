```python
def prepare_df(df):
    df["week"] = pd.to_numeric(df["week"])
    df["played"] = 1

    return df


def join_rank(base_dict: dict, add_dict: dict) -> dict:
    for key, value in add_dict.items():
        if key in base_dict:
            base_dict[key] += value
        else:
            base_dict[key] = value

    return base_dict


def get_top_3(
    team_season_df: pd.DataFrame,
    start: int,
    depth: int,
    count: int = 3,
    sort_field: str = "recAtt",
    rtn_field: str = "playerDisplay",
    blank_on_fail: bool = False,
) -> dict:
    search = team_season_df[
        (team_season_df["week"] <= start) & (team_season_df["week"] > start - depth)
    ]

    info = search.groupby(rtn_field).agg({"played": "count", sort_field: "sum"})

    info["att/g"] = info[sort_field] / info["played"]

    info = info.sort_values(by=["att/g"], ascending=False).head(3)

    if len(info.index) == 0:
        if blank_on_fail:
            return {}
        else:
            return get_top_3(
                team_season_df,
                start,
                depth,
                count,
                sort_field,
                rtn_field,
                blank_on_fail=True,
            )
    else:
        rtn = {}
        for i in range(min(count, len(info.index))):
            rtn[info.index[i]] = count - i

        return rtn


def get_rank(
    team_season_df, week, sort_field="recAtt", rtn_field="playerDisplay", count=3
):
    offset = week - 1
    res = join_rank(
        get_top_3(team_season_df, offset, 1, count, rtn_field),
        join_rank(
            get_top_3(team_season_df, offset, 3, count, rtn_field),
            get_top_3(team_season_df, offset, 5, count, rtn_field),
        ),
    )

    return dict(sorted(res.items(), key=lambda item: item[1], reverse=True))


def reduce_by_pos(season_df):
    usage = season_df.groupby("playerDisplay").agg(
        {"recAtt": "sum", "rushAtt": "sum", "playerPositionNorm": "last"}
    )

    rb_df = usage[
        (usage["recAtt"] > usage["rushAtt"]) & (usage["playerPositionNorm"] != "qb")
    ]

    targets = list(rb_df.index)

    return season_df[season_df["playerDisplay"].isin(targets)]


def run_team_analysis(team, season):
    season_df = reduce_by_pos(
        df[(df["teamDisplay"] == team) & (df["season"] == season)]
    )

    print("-----", team, season)
    print("week 16", get_rank(season_df, 16))
    print("week 15", get_rank(season_df, 15))
    print("week 14", get_rank(season_df, 14))
    print("week 13", get_rank(season_df, 13))
    print("week 9", get_rank(season_df, 9))
    print("week 5", get_rank(season_df, 5))
    print("week 3", get_rank(season_df, 3))


def run_test():
    run_team_analysis("PHI", "2023")
    run_team_analysis("PHI", "2021")
    run_team_analysis("PHI", "2022")
    run_team_analysis("PHI", "2020")
    run_team_analysis("PIT", "2023")
    run_team_analysis("PIT", "2022")
    run_team_analysis("PIT", "2021")
    run_team_analysis("PIT", "2020")

    run_team_analysis("KC", "2023")
    run_team_analysis("CAR", "2023")
    run_team_analysis("SEA", "2023")
    run_team_analysis("CIN", "2023")
    run_team_analysis("MIA", "2023")
    run_team_analysis("NE", "2023")


def run_league_analysis(season, week):
    teams = list(df["teamDisplay"].unique())
    season_df = reduce_by_pos(df[(df["season"] == int(season))])

    for team in teams:
        team_df = season_df[(season_df["teamDisplay"] == str(team))]

        print("team: ", team, " > week: " + str(week), get_rank(team_df, week))


run_league_analysis(2024, 3)
```
