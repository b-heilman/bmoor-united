```python
import os
import sys
import pandas as pd
import seaborn as sbn

pd.set_option('display.max_rows', 500)

# Get the directory of the current notebook
notebook_dir = os.path.dirname(os.path.abspath("__file__"))

# Add the parent directory to the system path
storage_path = os.path.abspath(notebook_dir + "/../../cache/parquet/base/games.parquet")

games_df = pd.read_parquet(storage_path).sort_values(
    by=["season", "week"], ascending=False
)

games_df['winningSince'] = games_df['winningSince'] / 60

games_df
```

```python
season = games_df[(games_df['season'] == 2024)]
```

```python
season.groupby(['winner'])['winningSince'].mean()
```

```python
sbn.displot(data=season, x="winningSince", hue="winner", kind="kde", cut=0)
```

```python
teams_season = season[(season['winner'].isin(['KC', 'PHI', 'DET', 'PIT', 'MIN']))]

sbn.displot(data=teams_season, x="winningSince", hue="winner", kind="kde", cut=0)
```

```python
team_averages = games_df.groupby(['season', 'winner'])['winningSince'].agg(('count', 'mean'))

print(team_averages)
```
